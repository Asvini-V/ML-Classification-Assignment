```python
import pickle
import pandas as pd
```

```python
loaded_model=pickle.load(open("finalized_model_LogReg.sav", 'rb'))
feature_columns = pickle.load(open("model_columns.pkl", "rb"))
```

```python
# Numerical inputs
age = float(input("Enter Age (age): "))
bp = float(input("Enter Blood Pressure in mm/Hg (bp): "))
al = float(input("Enter Albumin Level (al): "))
su = float(input("Enter Sugar Level (su): "))
bgr = float(input("Enter Blood Glucose Random (bgr): "))
bu = float(input("Enter Blood Urea (bu): "))
sc = float(input("Enter Serum Creatinine (sc): "))
sod = float(input("Enter Sodium (sod): "))
pot = float(input("Enter Potassium (pot): "))
hrmo = float(input("Enter Hemoglobin (hrmo): "))
pcv = float(input("Enter Packed Cell Volume (pcv): "))
wc = float(input("Enter White Blood Cell Count (wc): "))
rc = float(input("Enter Red Blood Cell Count (rc): "))

# Categorical inputs (0 or 1)
print("\nEnter 1 for YES / 0 for NO (or as specified)\n")

sg_b = int(input("Specific Gravity = b (sg_b): "))
sg_c = int(input("Specific Gravity = c (sg_c): "))
sg_d = int(input("Specific Gravity = d (sg_d): "))
sg_e = int(input("Specific Gravity = e (sg_e): "))

rbc_normal = int(input("Red Blood Cells Normal? (rbc_normal): "))
pc_normal = int(input("Pus Cell Normal? (pc_normal): "))
pcc_present = int(input("Pus Cell Clumps Present? (pcc_present): "))
ba_present = int(input("Bacteria Present? (ba_present): "))

htn_yes = int(input("Hypertension? (htn_yes): "))
dm_yes = int(input("Diabetes Mellitus? (dm_yes): "))
cad_yes = int(input("Coronary Artery Disease? (cad_yes): "))

appet_yes = int(input("Appetite Good? (appet_yes): "))
pe_yes = int(input("Pedal Edema? (pe_yes): "))
ane_yes = int(input("Anemia? (ane_yes): "))
```

```python
input_dict = {
    'age': age,
    'bp': bp,
    'al': al,
    'su': su,
    'bgr': bgr,
    'bu': bu,
    'sc': sc,
    'sod': sod,
    'pot': pot,
    'hrmo': hrmo,
    'pcv': pcv,
    'wc': wc,
    'rc': rc,
    'sg_b': sg_b,
    'sg_c': sg_c,
    'sg_d': sg_d,
    'sg_e': sg_e,
    'rbc_normal': rbc_normal,
    'pc_normal': pc_normal,
    'pcc_present': pcc_present,
    'ba_present': ba_present,
    'htn_yes': htn_yes,
    'dm_yes': dm_yes,
    'cad_yes': cad_yes,
    'appet_yes': appet_yes,
    'pe_yes': pe_yes,
    'ane_yes': ane_yes
}
```

```python
input_df = pd.DataFrame([input_dict])
```

```python
input_df = input_df[feature_columns]
```

```python
prediction = loaded_model.predict(input_df)
```

```python
 if prediction[0] == 1:
    print("\nðŸ”´ Result: Patient likely has Chronic Kidney Disease (CKD)")
else:
    print("\nðŸŸ¢ Result: Patient likely does NOT have CKD")
```
